```python
import pandas as pd

# Load datasets
customers = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_customers_dataset.csv')
orders = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_orders_dataset.csv')
order_items = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_order_items_dataset.csv')
payments = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_order_payments_dataset.csv')
reviews = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_order_reviews_dataset.csv')
products = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_products_dataset.csv')
sellers = pd.read_csv('/content/drive/My Drive/mini project/datasets/olist_sellers_dataset.csv')

# Merge customers with orders
merged_df = pd.merge(orders, customers, on='customer_id')

# Merge the above result with order items
merged_df = pd.merge(merged_df, order_items, on='order_id')

# Merge the above result with payments
merged_df = pd.merge(merged_df, payments, on='order_id')

# Merge the above result with reviews (left join as not all orders might have reviews)
merged_df = pd.merge(merged_df, reviews, on='order_id', how='left')

# Merge with products
merged_df = pd.merge(merged_df, products, on='product_id')

# Merge with sellers
merged_df = pd.merge(merged_df, sellers, on='seller_id')

# Preview the merged dataset
print(merged_df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix
from sklearn.decomposition import PCA

# Step 1: Load the Merged Dataset
merged_df = pd.read_csv('merged_dataset.csv')

# Step 2: Convert order_purchase_timestamp to datetime and create 'is_churn'
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
churn_threshold_date = merged_df['order_purchase_timestamp'].max() - pd.DateOffset(months=6)
merged_df['is_churn'] = (merged_df['order_purchase_timestamp'] < churn_threshold_date).astype(int)


X = merged_df.drop('is_churn', axis=1)
y = merged_df['is_churn']

# Identify categorical and numerical columns
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=['number']).columns.tolist()

# Initialize transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ]), numerical_features),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=True))  # Use sparse_output=True
        ]), categorical_features)
    ]
)

# Preprocess the features
X_preprocessed = preprocessor.fit_transform(X)

# Convert sparse matrix to CSR format for compatibility
X_preprocessed = csr_matrix(X_preprocessed)

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Step 5: Apply PCA (Optional)
pca = PCA(n_components=5)  # Keep 95% of variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Step 6: Implement a Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train_pca, y_train)
rf_predictions = rf_model.predict(X_test_pca)

# Evaluate the Random Forest model
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_predictions)}")
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# Step 7: Implement a Neural Network using TensorFlow/Keras
nn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_pca.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
nn_model.fit(X_train_pca, y_train, epochs=10, validation_data=(X_test_pca, y_test))

# Evaluate the Neural Network model
nn_loss, nn_accuracy = nn_model.evaluate(X_test_pca, y_test)


print(f"Neural Network Accuracy: {nn_accuracy}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NeighborhoodComponentsAnalysis
from scipy.sparse import csr_matrix
from sklearn.naive_bayes import BernoulliNB
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Step 1: Load the Merged Dataset
merged_df = pd.read_csv('merged_dataset.csv')

# Step 2: Convert order_purchase_timestamp to datetime and create 'is_churn'
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
churn_threshold_date = merged_df['order_purchase_timestamp'].max() - pd.DateOffset(months=6)
merged_df['is_churn'] = (merged_df['order_purchase_timestamp'] < churn_threshold_date).astype(int)

X = merged_df.drop('is_churn', axis=1)
y = merged_df['is_churn']

# Identify categorical and numerical columns
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=['number']).columns.tolist()

# Initialize transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', StandardScaler())
        ]), numerical_features),
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=True))  # Use sparse_output=True
        ]), categorical_features)
    ]
)

# Preprocess the features
X_preprocessed = preprocessor.fit_transform(X)

# Convert sparse matrix to CSR format for compatibility
X_preprocessed = csr_matrix(X_preprocessed)

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Step 5: Apply NCA
# Step 5: Apply NCA
nca = NeighborhoodComponentsAnalysis(n_components=5, random_state=42)  # Set number of components
X_train_nca = nca.fit_transform(X_train.toarray(), y_train) # Convert X_train to a dense array
X_test_nca = nca.transform(X_test.toarray()) # Convert X_test to a dense array

```

```python
# Step 6: Implement a Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train_nca, y_train)
rf_predictions = rf_model.predict(X_test_nca)

# Evaluate the Random Forest model
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_predictions)}")
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
```

```python

# Step 7: Implement a Neural Network using TensorFlow/Keras
nn_model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_nca.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
nn_model.fit(X_train_nca, y_train, epochs=10, validation_data=(X_test_nca, y_test))

# Evaluate the Neural Network model
nn_loss, nn_accuracy = nn_model.evaluate(X_test_nca, y_test)
print(f"Neural Network Accuracy: {nn_accuracy}")
```
